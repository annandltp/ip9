## use our data

```python
ids = data['Responder_id']
lang_response = data['LanguagesWorkedWith']

languange_counter = Counter()
```

```python
for response in lang_response:
    languange_counter.update(response.split(';'))
```

```python
print(languange_counter)
```

```python
print(languange_counter.most_common(15))
```

```python
language = []
popularity = []

for item in languange_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
```

```python
print(popularity)
```

## add to bar chart

```python
plt.bar(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Language')
plt.ylabel('Users')

plt.show()
```

```python
plt.barh(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Language')
# plt.ylabel('Users')

plt.show()
```

```python
language.reverse()
popularity.reverse()
plt.barh(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Number of People who Use')
# plt.ylabel('Users')

plt.show()
```
