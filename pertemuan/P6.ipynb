```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('data_sub.csv')
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot(ages, py_salaries, label="Python")
ax.plot(ages, js_salaries, label="JavaScript", color="m")
ax.plot(ages, dev_salaries, label="All Devs", linestyle="--")

ax.legend()
ax.set_title('Median Salary (USD) vs Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.show()
```

```python
fig, ax = plt.subplots(nrows= 2, ncols=1)
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1)
ax1.plot(ages, py_salaries, label="Python")

ax2.plot(ages, js_salaries, label="JavaScript", color="m")
ax2.plot(ages, dev_salaries, label="All Devs", linestyle="--")

ax1.legend()
ax1.set_title('Median Salary (USD) vs Age')
# ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
# ax2.set_title('Median Salary (USD) vs Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, sharex = True)
ax1.plot(ages, py_salaries, label="Python")

ax2.plot(ages, js_salaries, label="JavaScript", color="m")
ax2.plot(ages, dev_salaries, label="All Devs", linestyle="--")

ax1.legend()
ax1.set_title('Median Salary (USD) vs Age')
# ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
# ax2.set_title('Median Salary (USD) vs Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, sharey = True)
ax1.plot(ages, py_salaries, label="Python")

ax2.plot(ages, js_salaries, label="JavaScript", color="m")
ax2.plot(ages, dev_salaries, label="All Devs", linestyle="--")

ax1.legend()
# ax1.set_title('Median Salary (USD) vs Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
# ax2.set_title('Median Salary (USD) vs Age')
ax2.set_xlabel('Ages')
# ax2.set_ylabel('Median Salary (USD)')
fig.suptitle("Median Salary (USD) vs Age")
plt.tight_layout()
plt.savefig('subplot1.png')
plt.show()
```
